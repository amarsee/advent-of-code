## Part 1

- Each line of vents is given as a line segment in the format x1,y1 -> x2,y2 where x1,y1 are the coordinates of one end the line segment and x2,y2 are the coordinates of the other end. These line segments include the points at both ends. In other words:

    - An entry like 1,1 -> 1,3 covers points 1,1, 1,2, and 1,3.
    - An entry like 9,7 -> 7,7 covers points 9,7, 8,7, and 7,7.

```python
data = [x.strip() for x in open("input.txt").readlines()]
```

# Part 2

- Now consider diagonals as well

```python
df = pd.DataFrame(np.zeros((max_num+1, max_num+1)))
for line in data:
    nums = [int(x) for x in re.findall('\d+', line)]
    # x1 and x2 equal
    if nums[0] == nums[2]:
        df.loc[min([nums[1],nums[3]]):max([nums[1],nums[3]]), nums[0]] += 1
    # y1 and y2 equal
    elif nums[1] == nums[3]:
        df.loc[nums[1], min([nums[0],nums[2]]):max([nums[0],nums[2]])] += 1
    # diagonal
    elif (nums[3] - nums[1]) / (nums[2] - nums[0]) in [1, -1]:
        if (nums[3] - nums[1]) / (nums[2] - nums[0]) == -1:
            x = min([nums[0],nums[2]])
            y = max([nums[1],nums[3]])
            while x < max([nums[0],nums[2]])+1 :
                df.loc[y, x] += 1
                x += 1
                y -= 1
        else:
            x = min([nums[0],nums[2]])
            y = min([nums[1],nums[3]])
            while x < max([nums[0],nums[2]]) +1:
                df.loc[y, x] += 1
                x += 1
                y += 1

```

```python
running_count = 0
for col_name in df.columns:
    running_count += df.loc[df[col_name] > 1, col_name].shape[0]
```

```python
print(f"The output to Part 2 is {running_count}.")
```
